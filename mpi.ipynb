```c++
#include <stdio.h>
  
int size, rank;

MPI_Comm my_world = MPI_COMM_WORLD;

MPI_Comm_size(my_world, &size);
MPI_Comm_rank(my_world, &rank);

printf("I am rank %i out of %i!\n", rank, size);
```
